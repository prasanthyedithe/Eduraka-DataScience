```python
class Employee:
   'Common base class for all employees'
   empCount = 0

   def __init__(self, name, salary):
      self.name = name
      self.salary = salary
      Employee.empCount += 1
   
   def displayCount(self):
     print ("Total Employee %d" % Employee.empCount)

   def displayEmployee(self):
      print ("Name : ", self.name,  ", Salary: ", self.salary)
    
   def __del__(self):
      class_name = self.__class__.__name__
      print (class_name, "destroyed")

"This would create first object of Employee class"
emp1 = Employee("Zara", 2000)
"This would create second object of Employee class"
emp2 = Employee("Manni", 5000)
emp1.displayEmployee()
emp2.displayEmployee()
print ("Total Employee %d" % Employee.empCount)

hasattr(emp1, 'age') 
getattr(emp1, 'salary')   
setattr(emp1, 'salary', 8)
delattr(emp1, 'salary')   
delattr(emp1, 'name')

del emp1

print ("Total Employee %d" % Employee.empCount)
```

```python



# prices = {'infy':1450, 'tcs':3200, 'lntinfo':4600, 'wipro':2800, 'hcltech':1200}
def getPrices(stockList, mydate):
    prices = {}
    for stock in stockList:
        df = yfinance.download(stock, mydate)
        price = df.loc[mydate]['Close']
        prices[stock] = round(price,2)
    return prices

stockList = ['INFY.NS', 'TCS.NS', 'WIPRO.NS']
mydate = '2022-10-20'
prices = getPrices(stockList, mydate)
print(prices)
```

```python
import random
def createMovieMatrix(numUsers, numMovies):
    movie_matrix = []
    for user in range(numUsers):
        movies_rated_by_a_user = np.full(numMovies, -1)
        num_movies_rated = random.randint(0, numMovies-1)
        movies_that_user_rates = random.sample(range(numMovies), num_movies_rated)
        for i  in movies_that_user_rates:
            movies_rated_by_a_user[i] = random.randint(0, 10)
        movie_matrix.append(movies_rated_by_a_user)
    movie_matrix = np.array(movie_matrix)
    return movie_matrix

movie_matrix = createMovieMatrix(100, 1000)
expert_matrix = createMovieMatrix(10, 1000)
new_movie_matrix = createMovieMatrix(110, 50)
movie_matrix = np.vstack([movie_matrix, expert_matrix])
movie_matrix = np.hstack([movie_matrix, new_movie_matrix])
movie_matrix.shape
```

```python
# Joining NumPy Arrays
# There are 2 ways to join the Arrays using NumPy
# 1. concatenate
# 2. stack (hstack, vstack, dstack)

# 1-D
import numpy as np

arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.concatenate((arr1, arr2))

print(arr)

```

```python
# 2D
import numpy as np

arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[5, 6], [7, 8]])

arr = np.concatenate((arr1, arr2), axis=1)

print(arr)
```

```python
# 2. stack (hstack, vstack, dstack)
# Stacking is same as concatenation, the only difference is that stacking is done along a new axis.

import numpy as np

arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.stack((arr1, arr2), axis=1)

print(arr)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.hstack((arr1, arr2))

print(arr)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.vstack((arr1, arr2))

print(arr)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.dstack((arr1, arr2))

print(arr)
```
