### Type of Plots 


1.   Histrogram
2.   Bar Graph
3.   Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt

print(np.random.randn(1,2))
```

#### Histrogram

```python
y = np.random.randn(40,10)
plt.hist(y)

plt.grid(True)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Testing")

plt.show()
```

#### Bar chats

```python
import numpy as np
import matplotlib.pyplot as plt

plt.bar([1.5,2.6,5.1],[40,80,100])
plt.show()
```

```python
# xticks uses dictionary as input to show the bar charts with keys as lable

import matplotlib.pyplot as plt
import numpy as np

d = {"A":25,"B":70,"C":55,"D":90}

for i , key in enumerate(d):
  plt.bar(i,d[key])
plt.xticks(np.arange(len(d)),d.keys())
plt.show()
```

#### Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(3,4))
x= [40,20,5,40]
lable = ["A","B","C","D"]

plt.pie(x,labels=lable)
plt.show()
```

#### Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.random.randn(20)
y = np.random.randn(20)

plt.scatter(x,y)

plt.show()
```

### Style

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(1,3)

plt.plot(y, 'c')
plt.plot(y**2,'k')
plt.plot(y*2,'g')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(1,3)

plt.plot(y, '--')
plt.plot(y**2,':')
plt.plot(y*2,'-.')

plt.show()
```
